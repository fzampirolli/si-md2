## 🧪 Prática 2 — Tipos de Dados em Python

Python possui **tipagem dinâmica**: o tipo da variável é inferido
automaticamente. Conhecer os tipos básicos é essencial para entender
como os dados são representados internamente.

| Tipo     | Exemplo          | Uso comum em dados          |
|----------|------------------|-----------------------------|
| `int`    | `10`             | Contagens, IDs              |
| `float`  | `1.75`           | Medidas contínuas (altura)  |
| `str`    | `'Ana'`          | Nomes, categorias           |
| `bool`   | `True` / `False` | Flags, condições            |
| `list`   | `[1, 2, 3]`      | Sequências ordenadas        |
| `dict`   | `{'a': 1}`       | Dados estruturados (chave-valor) |

```python
# ----------------------------------------------------------
# Variáveis e tipos: cada linha demonstra um tipo diferente
# ----------------------------------------------------------
nome      = "Ana"       # str  — texto
peso_kg   = 55          # int  — inteiro
altura_m  = 1.65        # float — ponto flutuante
ativo     = True        # bool — booleano

# type() retorna o tipo da variável
variaveis = {"nome": nome, "peso_kg": peso_kg,
             "altura_m": altura_m, "ativo": ativo}

for chave, valor in variaveis.items():
    print(f"  {chave:<10} = {str(valor):<8}  tipo: {type(valor).__name__}")

# ----------------------------------------------------------
# F-strings: forma moderna e legível de formatar strings
# ----------------------------------------------------------
imc = peso_kg / altura_m ** 2
print(f"\nIMC de {nome}: {imc:.2f}  ← calculado em tempo real na f-string")
```

## 🧪 Prática 3 — Estruturas de Controle

As estruturas de controle (`if/elif/else`, `for`, `while`) permitem
que o programa tome decisões e repita operações — base de qualquer
algoritmo de Mineração de Dados.

> 💡 **Boas práticas:** Prefira `elif` a múltiplos `if` quando as
> condições são mutuamente exclusivas. Isso é mais eficiente e legível.

```python
# ----------------------------------------------------------
# Função com type hints e docstring padronizada (PEP 257)
# ----------------------------------------------------------
def classificar_imc(imc: float) -> str:
    """
    Classifica o estado nutricional com base no IMC (OMS).

    Args:
        imc: Índice de Massa Corporal (kg/m²).

    Returns:
        Classificação como string.
    """
    if imc < 18.5:
        return "Abaixo do peso"
    elif imc < 25.0:
        return "Normal"
    elif imc < 30.0:
        return "Sobrepeso"
    else:
        return "Obesidade"


# ----------------------------------------------------------
# Laço for + estrutura de dados (lista de dicionários)
# ----------------------------------------------------------
pacientes = [
    {"nome": "Ana",   "peso_kg": 55, "altura_m": 1.65},
    {"nome": "Bruno", "peso_kg": 95, "altura_m": 1.80},
    {"nome": "Carla", "peso_kg": 68, "altura_m": 1.70},
    {"nome": "David", "peso_kg": 82, "altura_m": 1.75},
]

print(f"{'Nome':<8} {'Peso':>6} {'Altura':>7} {'IMC':>6}  {'Status'}")
print("-" * 45)

for p in pacientes:
    imc    = p["peso_kg"] / p["altura_m"] ** 2
    status = classificar_imc(imc)
    print(f"{p['nome']:<8} {p['peso_kg']:>5}kg {p['altura_m']:>5}m "
          f"{imc:>6.2f}  {status}")
```

## 🧪 Prática 4 — List Comprehensions e Funções de Alta Ordem

**List comprehensions** são uma forma concisa e pythônica de criar
listas. São amplamente usadas em pipelines de dados.

As funções `map()`, `filter()` e `sorted()` são **funções de alta
ordem**: recebem outras funções como argumento — conceito fundamental
em programação funcional e muito usado em ciência de dados.

```python
# ----------------------------------------------------------
# List comprehension — equivale a um laço for + append
# ----------------------------------------------------------
imcs = [p["peso_kg"] / p["altura_m"] ** 2 for p in pacientes]
print("IMCs calculados:", [round(v, 2) for v in imcs])

# ----------------------------------------------------------
# filter() + lambda — seleciona elementos com condição
# ----------------------------------------------------------
em_atencao = list(filter(lambda imc: imc >= 25, imcs))
print(f"Casos que merecem atenção (IMC ≥ 25): {len(em_atencao)}")

# ----------------------------------------------------------
# sorted() com key customizada
# ----------------------------------------------------------
ordenados = sorted(pacientes, key=lambda p: p["peso_kg"] / p["altura_m"] ** 2)
print("\nPacientes ordenados por IMC (crescente):")
for p in ordenados:
    imc = p["peso_kg"] / p["altura_m"] ** 2
    print(f"  {p['nome']}: {imc:.2f}")
```

## 🧪 Prática 5 — Pandas: De Dados a Informação

O **Pandas** é a biblioteca central para manipulação de dados em Python.
Sua estrutura principal, o `DataFrame`, equivale a uma tabela relacional
e permite operações vetorizadas (sem laços explícitos), muito mais
eficientes para grandes volumes de dados.

> 💡 **Boas práticas:**
> - Prefira operações vetorizadas (`df['col'].apply(...)`) a laços `for`.
> - Use `df.copy()` ao modificar subconjuntos para evitar avisos de `SettingWithCopyWarning`.
> - Nomeie DataFrames de forma descritiva (`df_pacientes` em vez de apenas `df`).

```python
# ----------------------------------------------------------
# Criação do DataFrame a partir de lista de dicionários
# ----------------------------------------------------------
df_pacientes = pd.DataFrame(pacientes)

# Colunas calculadas — operação vetorizada (sem for!)
df_pacientes["imc"]    = df_pacientes["peso_kg"] / df_pacientes["altura_m"] ** 2
df_pacientes["status"] = df_pacientes["imc"].apply(classificar_imc)

# ----------------------------------------------------------
# Exploração básica — métodos essenciais do Pandas
# ----------------------------------------------------------
print("=== DataFrame completo ===")
print(df_pacientes.round(2).to_string(index=False))

print("\n=== Estatísticas descritivas ===")
print(df_pacientes[["peso_kg", "altura_m", "imc"]].describe().round(2))

print("\n=== Filtragem: pacientes com IMC ≥ 25 ===")
print(df_pacientes[df_pacientes["imc"] >= 25][["nome", "imc", "status"]].to_string(index=False))
```

## 🧪 Prática 6 — Visualização: Do Dado ao Conhecimento

A visualização é a etapa de **pós-processamento** do KDD: transforma
dados e informações em representações visuais que facilitam a
interpretação e a tomada de decisão.

> 💡 **Boas práticas de visualização:**
> - Sempre rotule eixos e adicione título.
> - Use cores com significado semântico (vermelho = alerta, verde = ok).
> - Salve figuras com `plt.savefig()` para relatórios.
> - Chame `plt.tight_layout()` para evitar sobreposição de elementos.

```python
# ----------------------------------------------------------
# Mapa de cores semântico baseado no status
# ----------------------------------------------------------
mapa_cores = {
    "Abaixo do peso": "#3498db",  # azul
    "Normal":         "#2ecc71",  # verde
    "Sobrepeso":      "#f39c12",  # laranja
    "Obesidade":      "#e74c3c",  # vermelho
}
cores = df_pacientes["status"].map(mapa_cores)

# ----------------------------------------------------------
# Figura com dois subplots: barras + dispersão
# ----------------------------------------------------------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
fig.suptitle("Análise de Saúde dos Pacientes — Hierarquia DIKW", fontsize=13)

# --- Subplot 1: IMC por paciente (barras) ---
ax1.bar(df_pacientes["nome"], df_pacientes["imc"], color=cores, edgecolor="white")
ax1.axhline(y=25, color="red",    linestyle="--", linewidth=1.2, label="Sobrepeso (25)")
ax1.axhline(y=18.5, color="blue", linestyle=":",  linewidth=1.2, label="Abaixo do peso (18.5)")
ax1.set_title("IMC por Paciente (Informação)")
ax1.set_ylabel("IMC (kg/m²)")
ax1.set_ylim(0, 35)
ax1.legend(fontsize=8)

# Anota o valor em cima de cada barra
for i, (nome, imc) in enumerate(zip(df_pacientes["nome"], df_pacientes["imc"])):
    ax1.text(i, imc + 0.4, f"{imc:.1f}", ha="center", fontsize=9)

# --- Subplot 2: Peso x Altura (dispersão) ---
scatter = ax2.scatter(
    df_pacientes["altura_m"],
    df_pacientes["peso_kg"],
    c=df_pacientes["imc"],
    cmap="RdYlGn_r",
    s=200,
    edgecolors="gray",
    zorder=3
)
for _, row in df_pacientes.iterrows():
    ax2.annotate(row["nome"], (row["altura_m"], row["peso_kg"]),
                 textcoords="offset points", xytext=(6, 4), fontsize=9)

plt.colorbar(scatter, ax=ax2, label="IMC")
ax2.set_title("Peso vs. Altura (Conhecimento)")
ax2.set_xlabel("Altura (m)")
ax2.set_ylabel("Peso (kg)")
ax2.grid(True, linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()

print("""
RELATÓRIO — Hierarquia DIKW aplicada:
  DADO       → peso e altura de cada paciente
  INFORMAÇÃO → IMC calculado
  CONHECIMENTO → classificação e identificação de casos de atenção
  SABEDORIA  → decisão clínica informada pelo profissional de saúde
""")
```

## 🧪 Prática 7 — Modelagem Preditiva Simples (KDD na Prática)

Neste exemplo avançado, utilizamos a hierarquia KDD completa:
dados brutos → pré-processamento → "mineração" (ajuste de função)
→ pós-processamento (predição e visualização).

Modelamos a equação do movimento uniformemente acelerado:

$$d = \frac{1}{2} g \cdot t^2$$

O sistema "descobre" o padrão quadrático e usa-o para **prever** a
posição em instantes não observados — isso é **modelagem preditiva**.

```python
import numpy as np

# ----------------------------------------------------------
# 1. DADOS — observações brutas (com ruído realista)
# ----------------------------------------------------------
rng  = np.random.default_rng(seed=42)        # semente para reprodutibilidade
g    = 9.8                                    # aceleração da gravidade (m/s²)
t_obs = np.linspace(0, 5, 12)                # 12 medições de 0 a 5 s
d_obs = 0.5 * g * t_obs**2 + rng.normal(0, 2, len(t_obs))  # + ruído

# ----------------------------------------------------------
# 2. MINERAÇÃO — ajuste polinomial de grau 2 (np.polyfit)
# ----------------------------------------------------------
coefs   = np.polyfit(t_obs, d_obs, deg=2)    # encontra os coeficientes
modelo  = np.poly1d(coefs)                   # cria função polinomial

print("Coeficientes descobertos (ax² + bx + c):")
print(f"  a = {coefs[0]:.3f}  (esperado ≈ {0.5*g:.3f} = g/2)")
print(f"  b = {coefs[1]:.3f}  (esperado ≈ 0)")
print(f"  c = {coefs[2]:.3f}  (esperado ≈ 0)")

# ----------------------------------------------------------
# 3. PREDIÇÃO — estima distância para t=6 e t=7 (não observados)
# ----------------------------------------------------------
t_futuro   = np.array([6.0, 7.0])
d_previsto = modelo(t_futuro)
d_real     = 0.5 * g * t_futuro**2

print("\nCapacidade preditiva:")
for t, dp, dr in zip(t_futuro, d_previsto, d_real):
    erro = abs(dp - dr)
    print(f"  t={t:.0f}s → previsto={dp:.1f}m | real={dr:.1f}m | erro={erro:.2f}m")

# ----------------------------------------------------------
# 4. VISUALIZAÇÃO
# ----------------------------------------------------------
t_curva = np.linspace(0, 7.5, 200)

plt.figure(figsize=(8, 4))
plt.scatter(t_obs, d_obs, color="steelblue", zorder=5, label="Observações (com ruído)")
plt.plot(t_curva, modelo(t_curva), color="tomato", linewidth=2, label="Modelo descoberto")
plt.plot(t_curva, 0.5*g*t_curva**2, color="green", linewidth=1.5,
         linestyle="--", label="Curva real (d = g·t²/2)")
plt.scatter(t_futuro, d_previsto, marker="*", s=200, color="gold",
            edgecolors="black", zorder=6, label="Predições (t=6, 7 s)")
plt.title("KDD Aplicado: Queda Livre — Descoberta de Padrão e Predição")
plt.xlabel("Tempo (s)")
plt.ylabel("Distância percorrida (m)")
plt.legend(fontsize=8)
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```
