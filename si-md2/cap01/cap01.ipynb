### üß™ Pr√°tica: Transformando Dado em Informa√ß√£o (IMC)


Na biblioteca **Pandas**, o **DataFrame** constitui a estrutura de dados central. Trata-se de uma tabela bidimensional ‚Äî an√°loga a uma planilha de c√°lculo ou a uma tabela de banco de dados ‚Äî na qual os dados s√£o organizados em **linhas** e **colunas**.

No exemplo apresentado na @fig-1-imc-analise, o c√≥digo executa as seguintes opera√ß√µes:

* **`pd.DataFrame(...)`**: O construtor da biblioteca Pandas √© invocado para a cria√ß√£o de uma nova tabela.
* **`dados_pacientes`**: O dicion√°rio contendo nomes, pesos e alturas √© utilizado como a fonte de dados prim√°ria.
* **Mapeamento**: As **chaves** do dicion√°rio (`'Paciente'`, `'Peso_kg'`, `'Altura_m'`) s√£o automaticamente convertidas em **cabe√ßalhos de colunas**, enquanto as **listas de valores** passam a compor as **linhas** da tabela.
* **`df`**: A tabela resultante √© armazenada na vari√°vel `df` (abrevia√ß√£o convencional para *DataFrame*), permitindo a execu√ß√£o de c√°lculos complexos e filtragens de alto desempenho.


```python
#| quarto-raw: true
#| label: fig-1-imc-analise
#| fig-cap: "An√°lise de IMC e classifica√ß√£o de sa√∫de dos pacientes."

import pandas as pd
import matplotlib.pyplot as plt

"""
1. DADOS BRUTOS (Data):
Fatos registrados sem contexto. Isolados, n√£o permitem tomada de decis√£o.
"""
dados_pacientes = {
    'Paciente': ['Ana', 'Bruno', 'Carla', 'David'],
    'Peso_kg': [55, 95, 68, 82],
    'Altura_m': [1.65, 1.80, 1.70, 1.75]
}

df = pd.DataFrame(dados_pacientes)

"""
2. TRANSFORMA√á√ÉO EM INFORMA√á√ÉO (Information):
O c√°lculo do IMC contextualiza peso e altura, conferindo significado aos dados.
"""
df['IMC'] = df['Peso_kg'] / (df['Altura_m'] ** 2)

# 3. RUMO AO CONHECIMENTO (Knowledge):
# Aplicamos uma regra de neg√≥cio (Classifica√ß√£o) para extrair padr√µes.
def classificar_saude(imc):
    if imc < 18.5: return 'Abaixo do peso'
    elif imc < 25: return 'Normal'
    else: return 'Sobrepeso'

df['Status'] = df['IMC'].apply(classificar_saude)

print("=== Tabela Processada (Informa√ß√£o) ===")
print(df.round(2))

# 4. VISUALIZA√á√ÉO PARA TOMADA DE DECIS√ÉO (Intelig√™ncia/A√ß√£o)
plt.figure(figsize=(7, 4))
cores = ['skyblue' if s == 'Normal' else 'salmon' for s in df['Status']]
plt.bar(df['Paciente'], df['IMC'], color=cores)
plt.axhline(y=25, color='red', linestyle='--', label='Limite Normalidade (OMS)')

plt.title('An√°lise de Sa√∫de dos Pacientes')
plt.ylabel('√çndice de Massa Corporal (IMC)')
plt.legend()
plt.show()

print(f"""
RELAT√ìRIO DE EXECU√á√ÉO:
======================
- M√©dia de IMC: {df['IMC'].mean():.2f}
- Casos de Aten√ß√£o: {len(df[df['Status'] != 'Normal'])}
- Status do ambiente: PRONTO.
""")
```

A utiliza√ß√£o do comando `pd.DataFrame(dados_pacientes)` representa o est√°gio inicial do processamento de dados. Este procedimento converte estruturas est√°ticas em objetos din√¢micos aptos √† manipula√ß√£o computacional.

O exemplo ilustra a transi√ß√£o na hierarquia **DIKW** (Dado, Informa√ß√£o, Conhecimento e Sabedoria). Os **Dados** brutos s√£o estruturados e processados para a gera√ß√£o de **Informa√ß√£o** (IMC). Subsequentemente, emprega-se a tarefa de **Classifica√ß√£o** ‚Äî pilar da Minera√ß√£o de Dados ‚Äî para a rotulagem do estado de sa√∫de, transformando a informa√ß√£o em **Conhecimento** pass√≠vel de aplica√ß√£o em processos de **Tomada de Decis√£o**.
